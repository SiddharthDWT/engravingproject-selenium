```python
from selenium import webdriver
driver=webdriver.Chrome("chromedriver.exe")
```

```python
#https://www.pwengraving.com/sports/soccer/?limit=all manual entry for particular pager
url="https://www.pwengraving.com/sports/soccer/?limit=all" #https://www.pwengraving.com/sports/soccer/?limit=all manual entry for particular pager
driver.get(url)
```

```python
%%timeit


Urls_soccer=[]
wait = WebDriverWait(driver,2)
Url_links=driver.find_elements("xpath","//li[@class='item first col-xs-12 col-sm-4' or @class='item last col-xs-12 col-sm-4']/a")
for i in Url_links:
    try:
        Urls_soccer.append(i.get_attribute("href"))
    except:
        Urls_soccer.append("-")
```

```python
len(Urls_soccer)
```

```python
#  url1=Urls_soccer[0:5]
```

```python
url1
```

```python

h1_tag=[]
meta_discription=[]
price=[]
title=[]

for i in Urls_soccer:
    driver.get(i)
    time.sleep(1)
# find complete title of page

    try:
        titles=driver.title
        title.append(titles)
    except:
        title.append("-")
    
# xpath element for h1-tag    
    try:
        h1=driver.find_element("xpath",'//h1')
        h1_tag.append(h1.text.replace("'",""))
    except:
        h1_tag.append("-")
#xpath element for meta discription
        
    try:
        meta=driver.find_element("xpath",'//div[@class="std"]')
        meta_discription.append(meta.text.replace("\n"," "))
    except:
        meta_discription.append("-")
        
# xpath element for price attribute

    try:
        prices=driver.find_element("xpath",'//span[@class="regular-price"]/span')
        price.append(prices.text.replace("$",""))
        
    except:
        price.append("-")
   
print(len(Urls_soccer))
print(len(h1_tag))
print(len(price))
print(len(meta_discription))
print(len(title))
```

```python
import pandas as pd
df1=pd.DataFrame({})
df1['Urls']=Urls_soccer
df1['title']=title
df1['h1_tag']=h1_tag
df1['price']=price
df1['meta_discription']=meta_discription

Soccer_data=df1
```

```python
# writing to Excel
datatoexcel1 = pd.ExcelWriter('Soccer_data.xlsx')
# write DataFrame to excel
Soccer_data.to_excel(datatoexcel1)
# save the excel
datatoexcel1.save()
print('DataFrame1 is written to Excel File successfully.')
```
